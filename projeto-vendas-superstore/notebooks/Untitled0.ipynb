```python
# Baixar o dataset direto da internet
import pandas as pd

# URL do dataset Superstore - link original
# url = 'https://raw.githubusercontent.com/plotly/datasets/master/superstore.csv'

# Carregar o dataset localmente, pois o link original pode estar quebrado ou o arquivo foi renomeado/removido
df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')

# Verificar se deu certo
print("✅ Dataset carregado com sucesso!")
print(f"Total de linhas: {len(df)}")
print(f"Total de colunas: {len(df.columns)}")
print("\nPrimeiras linhas:")
df.head()
```

```python
# Ver informações gerais
print("📊 Informações do Dataset:\n")
df.info()

print("\n" + "="*50)
print("📋 Colunas disponíveis:")
print("="*50)
for i, col in enumerate(df.columns, 1):
    print(f"{i}. {col}")

print("\n" + "="*50)
print("🔍 Valores nulos:")
print("="*50)
print(df.isnull().sum())

print("\n" + "="*50)
print("📈 Estatísticas descritivas:")
print("="*50)
df.describe()
```

```python
# PREPARAÇÃO E TRANSFORMAÇÃO DOS DADOS

# Converter datas
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Criar colunas de tempo
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Month Name'] = df['Order Date'].dt.strftime('%B')
df['Quarter'] = df['Order Date'].dt.quarter
df['Day of Week'] = df['Order Date'].dt.day_name()
df['Week Number'] = df['Order Date'].dt.isocalendar().week

# Métricas calculadas
df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days
df['Profit Margin'] = (df['Profit'] / df['Sales']) * 100
df['Revenue per Item'] = df['Sales'] / df['Quantity']

# Categorizar margem de lucro
def categorize_margin(margin):
    if margin < 0:
        return 'Prejuízo'
    elif margin < 10:
        return 'Baixa'
    elif margin < 30:
        return 'Média'
    else:
        return 'Alta'

df['Margin Category'] = df['Profit Margin'].apply(categorize_margin)

print("✅ Transformações concluídas!")
print(f"\nPeríodo dos dados: {df['Order Date'].min().date()} até {df['Order Date'].max().date()}")
print(f"Total de anos: {df['Year'].nunique()}")
print(f"\nNovas colunas criadas:")
print(df.columns[-10:].tolist())
```

```python
# ANÁLISE RFM - Recency, Frequency, Monetary

# Data de referência (última venda + 1 dia)
reference_date = df['Order Date'].max() + pd.Timedelta(days=1)

# Calcular RFM por cliente
rfm = df.groupby('Customer ID').agg({
    'Order Date': lambda x: (reference_date - x.max()).days,  # Recency
    'Order ID': 'nunique',  # Frequency
    'Sales': 'sum'  # Monetary
}).reset_index()

rfm.columns = ['Customer ID', 'Recency', 'Frequency', 'Monetary']

# Criar quartis para pontuação (1-4)
rfm['R_Score'] = pd.qcut(rfm['Recency'], 4, labels=[4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels=[1,2,3,4])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 4, labels=[1,2,3,4])

# Score RFM combinado
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

# Segmentação de clientes
def segment_customer(row):
    r, f, m = int(row['R_Score']), int(row['F_Score']), int(row['M_Score'])

    if r >= 3 and f >= 3 and m >= 3:
        return 'Champions'
    elif r >= 3 and f >= 2:
        return 'Loyal Customers'
    elif r >= 3 and f == 1:
        return 'Potential Loyalists'
    elif r <= 2 and f >= 3:
        return 'At Risk'
    elif r <= 2 and f <= 2:
        return 'Hibernating'
    else:
        return 'Need Attention'

rfm['Customer Segment'] = rfm.apply(segment_customer, axis=1)

# Juntar com dados originais
df = df.merge(rfm[['Customer ID', 'Customer Segment', 'RFM_Score']], on='Customer ID', how='left')

print("✅ Análise RFM concluída!")
print(f"\nTotal de clientes: {rfm['Customer ID'].nunique()}")
print("\n📊 Distribuição de Segmentos:")
print(rfm['Customer Segment'].value_counts())
print("\n" + "="*50)
print(rfm.head(10))
```

```python
# CRIAR TABELAS AGREGADAS PARA O POWER BI

# 1. Tabela de Vendas por Período
vendas_tempo = df.groupby(['Year', 'Quarter', 'Month', 'Month Name']).agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum',
    'Order ID': 'nunique',
    'Customer ID': 'nunique'
}).reset_index()

vendas_tempo.columns = ['Year', 'Quarter', 'Month', 'Month Name',
                        'Total Sales', 'Total Profit', 'Total Quantity',
                        'Total Orders', 'Unique Customers']
vendas_tempo['Profit Margin'] = (vendas_tempo['Total Profit'] / vendas_tempo['Total Sales']) * 100
vendas_tempo['Average Order Value'] = vendas_tempo['Total Sales'] / vendas_tempo['Total Orders']

# 2. Tabela de Produtos
produtos = df.groupby(['Category', 'Sub-Category', 'Product Name']).agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum',
    'Order ID': 'nunique'
}).reset_index()

produtos.columns = ['Category', 'Sub-Category', 'Product Name',
                   'Total Sales', 'Total Profit', 'Total Quantity', 'Total Orders']
produtos['Profit Margin'] = (produtos['Total Profit'] / produtos['Total Sales']) * 100

# 3. Tabela de Clientes
clientes = df.groupby('Customer ID').agg({
    'Customer Name': 'first',
    'Segment': 'first',
    'Customer Segment': 'first',
    'State': 'first',
    'Region': 'first',
    'Sales': 'sum',
    'Profit': 'sum',
    'Order ID': 'nunique',
    'Order Date': ['min', 'max']
}).reset_index()

clientes.columns = ['Customer ID', 'Customer Name', 'Segment', 'Customer Segment',
                   'State', 'Region', 'Total Sales', 'Total Profit',
                   'Total Orders', 'First Purchase', 'Last Purchase']
clientes['Customer Lifetime Days'] = (clientes['Last Purchase'] - clientes['First Purchase']).dt.days
clientes['Average Order Value'] = clientes['Total Sales'] / clientes['Total Orders']

# 4. Tabela Geográfica
geografia = df.groupby(['Country', 'Region', 'State', 'City']).agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Order ID': 'nunique',
    'Customer ID': 'nunique'
}).reset_index()

geografia.columns = ['Country', 'Region', 'State', 'City',
                    'Total Sales', 'Total Profit', 'Total Orders', 'Total Customers']

# Métricas gerais
total_vendas = df['Sales'].sum()
total_lucro = df['Profit'].sum()
margem_media = (total_lucro / total_vendas) * 100

print("Tabelas agregadas criadas com sucesso!")
print("\nResumo das tabelas:")
print("- Vendas por Tempo:", len(vendas_tempo), "linhas")
print("- Produtos:", len(produtos), "linhas")
print("- Clientes:", len(clientes), "linhas")
print("- Geografia:", len(geografia), "linhas")
print("\nPrincipais insights:")
print("- Vendas totais: $", round(total_vendas, 2))
print("- Lucro total: $", round(total_lucro, 2))
print("- Margem media:", round(margem_media, 2), "%")
```
